# 03_satellite.ipynb — NASA HLS S30 NDVI Extraction

**Executable version runs on SageMaker. Obtain output CSV from teammate or SageMaker environment.**

---

## Spec

- Queries NASA CMR STAC (`HLSS30.v2.0`) via `earthaccess` + `pystac-client`
- Pulls NIR (B8A) and Red (B04) per state bounding box per forecast date (±7 day window)
- Filters to <20% cloud cover; caps at 6 tiles per window
- Computes per-tile NDVI = (NIR − Red) / (NIR + Red), averages across tiles
- Reshapes to wide format and forward-fills cloudy windows via interpolation
- Output: `data/raw/ndvi_by_state_date.csv` — columns: `state`, `year`, `ndvi_aug1`, `ndvi_sep1`, `ndvi_oct1`, `ndvi_final`
- Runs on SageMaker (IAM role provides Earthdata auth via `.netrc`)
- Status: Iowa complete, Colorado in progress, remaining states pending
