```python
# ── QuickStats (tabular) ──────────────────────────────────────────────
import pandas as pd
import matplotlib.pyplot as plt

# Replace with actual path/URL when provided
qs = pd.read_csv("quickstats.csv")
print(qs.head())
print(qs.columns.tolist())

qs.groupby("state_name")["Value"].sum().sort_values().plot(
    kind="barh", figsize=(10, 8), title="QuickStats — Value by State"
)
plt.tight_layout()
plt.show()
```
